```python
#Lets start with order items analysis
#Aggregate per item statistics
df = fct_order_items.groupby(fct_order_items["item_id"])
per_item_stats = df[["quantity", "price", "cost"]].agg(['sum', 'mean', 'max', 'min'])
per_item_stats.head()
```

```python
#Cost > Price, which means cost is for the company to sell the item at, and price is how much it costs the company to procure/make the item.
```

```python
#Scatter plot of price vs cost (Remove outliers for better visualization)
x = per_item_stats['price'].quantile([0.05,0.50])
y = per_item_stats['cost'].quantile([0.05,0.50])
plt.figure(figsize=(10,6))
plt.scatter(x, y, alpha=0.5)
plt.title('Price vs Cost Scatter Plot')
plt.xlabel('Price')
plt.ylabel('Cost')
plt.show()
```

```python
#Time-series analysis of order items sold over time
#See how 10 items perform over time
top10_items = (fct_order_items.groupby('item_id')['quantity']
               .sum()
               .sort_values(ascending=False)
               .head(10)
               .index)

filtered = fct_order_items[fct_order_items["item_id"].isin(top10_items)].copy()
filtered = filtered.reset_index()

# # weekly totals per item
# qty_by_week = (
#     filtered.groupby(["item_id", pd.Grouper(key="created", freq="W")])["quantity"]
#     .sum()
#     .reset_index()
#     .sort_values(["created", "item_id"])
# )

# plt.figure(figsize=(14, 6))
# sns.lineplot(data=qty_by_week, x="created", y="quantity", hue="item_id")
# plt.title("Top 10 item_ids: quantity sold over time (weekly)")
# plt.xlabel("Week")
# plt.ylabel("Quantity sold")
# plt.legend(title="item_id", bbox_to_anchor=(1.02, 1), loc="upper left")
# plt.tight_layout()
# plt.show()

qty_by_day = (
    filtered.groupby(["item_id", pd.Grouper(key="created", freq="D")])["quantity"]
    .sum()
    .reset_index()
    .sort_values(by=["created", "item_id"])
)

plt.figure(figsize=(14, 6))
sns.lineplot(data=qty_by_day, x="created", y="quantity", hue="item_id", marker="o")
plt.title("Top 10 item_ids: quantity sold over time (daily)")
plt.xlabel("Created date")
plt.ylabel("Quantity sold")
plt.legend(title="item_id", bbox_to_anchor=(1.02, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

```python
qty_by_day.sort_values("quantity", ascending=False).head(10)
```

```python
qty_by_day2 = qty_by_day.sort_values(["item_id", "created"]).copy()
qty_by_day2["qty_7d_ma"] = qty_by_day2.groupby("item_id")["quantity"].transform(
    lambda s: s.rolling(7, min_periods=1).mean()
)

plt.figure(figsize=(14, 6))
sns.lineplot(data=qty_by_day2, x="created", y="qty_7d_ma", hue="item_id")
plt.title("Daily quantity (7-day moving average)")
plt.ylabel("Quantity sold (7D MA)")
plt.tight_layout()
plt.show()
```
